```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
# import metpy.calc as mpcalc
import numpy as np
import xarray as xr
import cmocean.cm as cmo
from scipy.stats import pearsonr
```

```python
assi = np.load("/work/uo1075/u241321/data/assi_eemd_t_10.npy")[2,:,:] # mode(input, c1-c4), box, time
na = np.load("/work/uo1075/u241321/data/eemd_NAO_DJF.npy")   #  
nao = na[2,:]
```

```python
# # Specify longitude values for chosen domain
blat= [55, 50, 52.5, 55, 58, 61, 58.5, 63.5, 68.5, 76.5]
blon= [-51, -48, -34.5, -24.5, -16, -2.5, 3.5, 3.5, 11.5, 10.5]
```

```python
n=10  # 10 stations
corr = np.zeros((n))

for i in range(n):
        corr[i]=pearsonr(assi[i,:], nao)[0]
```

```python

x = np.arange(1,11,1)


fig, ax = plt.subplots(figsize = [15,10])

# colors=black
ax.bar(x, corr, color='darkslategrey')


fig.suptitle('(b) correlation: NAO & Temperature along currents', x=0.495, y=0.93, fontsize=32)
plt.xlabel('Locations along ocean currents', fontsize=35)
plt.ylabel('Corr', fontsize=30)
ax.set_xticks(np.arange(1,11,1))
ax.set_xticklabels(np.arange(1,11,1), fontsize=35)
ax.tick_params(axis='both', labelsize=40)
# ax.set(xlabel='Along ocean currents ', ylabel='Corr ',
#        title='(b) correlation: NAO & Temperature along currents')
# ax.set_ylim(-1, 0.8)
# ax.grid()
# ax.legend(loc='upper right', fontsize='16')
# ax.annotate('5%', fontsize='16',
#             xy=(80, 150), xycoords='figure points')

plt.rcParams["font.size"] = "25"
plt.show()
fig.savefig('/work/uo1075/u241321/results/p7.2.pdf', bbox_inches='tight')    
```
