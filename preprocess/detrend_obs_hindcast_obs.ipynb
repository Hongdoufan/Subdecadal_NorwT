```python
import xarray as xr
import numpy as np
from xarray import DataArray
import pandas as pd
from scipy import signal
```

```python
data1       = xr.open_dataset("/work/uo1075/u241321/data_cdo/Obs/EN4_s_ym_remap_NA560.nc") 
data2       = xr.open_dataset("/work/uo1075/u241321/data_cdo/Obs/EN4_t_ym_remap_NA560.nc") 
data3      = xr.open_dataset("/work/uo1075/u241321/data_cdo/so_ly_1969-2019_hindcast.nc",decode_times=False) 
data4      = xr.open_dataset("/work/uo1075/u241321/data_cdo/thetao_ly_1969-2019_hindcast.nc",decode_times=False) 


# obs anomalies for 1970-2019
s_o = data1 ['salinity']
s_oo = s_o[9:,:,:]-np.mean(s_o[9:,:,:],axis=0)

t_o = data2 ['temperature']
t_oo = t_o[9:,:,:]-np.mean(t_o[9:,:,:],axis=0)
```

```python
dropped =s_oo.stack(feature=("depth","lat","lon")).dropna(dim="feature")
detrend = signal.detrend(dropped ,axis=0)
feature = dropped .coords["feature"]
time = dropped .coords["time"]
detrend = xr.DataArray(detrend, dims = ["time","feature"], coords = {"time":time,"feature":feature}).unstack()
salinity = detrend
s_oo.stack(feature=("depth","lat","lon")).dropna(dim="feature").unstack()
salinity.to_netcdf("/work/uo1075/u241321/data_cdo/Obs/EN4_s_ym_remap_NA560_dt.nc")
```

```python
dropped =t_oo.stack(feature=("depth","lat","lon")).dropna(dim="feature")
detrend = signal.detrend(dropped ,axis=0)
feature = dropped .coords["feature"]
time = dropped .coords["time"]
detrend = xr.DataArray(detrend, dims = ["time","feature"], coords = {"time":time,"feature":feature}).unstack()
temperature = detrend

temperature.to_netcdf("/work/uo1075/u241321/data_cdo/Obs/EN4_t_ym_remap_NA560_dt.nc")
```

```python
s_h = data3['so']
dropped =s_h.stack(feature=("member","lead_year","depth","lat","lon")).dropna(dim="feature")
detrend = signal.detrend(dropped ,axis=0)
feature = dropped .coords["feature"]
time = dropped .coords["time"]
detrend = xr.DataArray(detrend, dims = ["time","feature"], coords = {"time":time,"feature":feature}).unstack()
salinity = detrend

salinity.to_netcdf("/work/uo1075/u241321/data_cdo/salinity_ly_1969-2019_hindcast_dt.nc")
```

```python
t_h = data4['thetao']
dropped =t_h.stack(feature=("member","lead_year","depth","lat","lon")).dropna(dim="feature")
detrend = signal.detrend(dropped ,axis=0)
feature = dropped .coords["feature"]
time = dropped .coords["time"]
detrend = xr.DataArray(detrend, dims = ["time","feature"], coords = {"time":time,"feature":feature}).unstack()
temperature = detrend

temperature.to_netcdf("/work/uo1075/u241321/data_cdo/temperature_ly_1969-2019_hindcast_dt.nc")
```
